```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def theta(r,a,b,D):
    theta = (a/b)*np.log(r) + D
    return theta
```

```python
def xy(r,theta):
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    return(x,y)
```

```python
def v_inds(inds,r,theta,vr,vth):
    
    x_i = r[inds]*np.cos(theta[inds])
    y_i = r[inds]*np.sin(theta[inds])
    vr_i = vr[inds]
    vth_i = vth[inds]
    vx_i = np.cos(theta[inds])*vr_i - np.sin(theta[inds])*vth_i
    vy_i = np.sin(theta[inds])*vr_i + np.cos(theta[inds])*vth_i
    
    return(x_i,y_i,vx_i,vy_i)
```

```python
#4b)
init = [1.,0.]
xi = np.linspace(0,20,1000)
n = 0
```

```python
def ds(coords, xi, n):
    theta,u = coords
    du_dxi = (-xi**2)*theta**n
    if xi == 0:
        d_th_dxi = 0
    else:
        d_th_dxi = u/(xi**2)
    dcoords = [d_th_dxi,du_dxi]
    return dcoords
```

```python
n=0
sol1= odeint(ds, init, xi, args=(n,)) # solve
theta1= 1 - (xi**2/(6))
plt.plot(xi, theta1,linestyle = '--')
plt.plot(xi,sol1[:,0])
```
